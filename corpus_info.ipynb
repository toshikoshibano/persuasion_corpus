## Getting Started

```python
import time
import numpy as np
import pandas as pd
import pickle
import os.path as path
import csv
import random, os, json, csv
import nltk
nltk.download('stopwords', quiet=True) | nltk.download('punkt', quiet=True)
from nltk import sent_tokenize, word_tokenize
from collections import Counter
```

```python
def load_master(master_dir):
    """Returns three master files present in `master_dir`. If such directory does
    not exist, create one and three master files, and then return them. 
    """    
    with open(os.path.join(master_dir, 'submission.json'), 'r', encoding='utf-8') as f:
        submission_dict = json.load(f)
    with open(os.path.join(master_dir, 'delta_thread.json'), 'r', encoding='utf-8') as f:
        delta_thread_dict = json.load(f)
    with open(os.path.join(master_dir, 'log.json'), 'r') as f:
        log_dict = json.load(f)

    return submission_dict, delta_thread_dict, log_dict
```

```python
MASTER_DIR = 'reddit_corpus'  #'mar4'  #'feb26'
submission_dict, delta_thread_dict, log_dict = load_master(MASTER_DIR)
```

```python
print(len(submission_dict['submission_id']), "submission posts in total")
print(len(delta_thread_dict.keys()), "threads in total")
```

```python
def get_title_word_counts(submission_id):
    idx = submission_dict['submission_id'].index(submission_id)
    words = word_tokenize(submission_dict['submission_title'][idx])
    return len(words)
```

```python
def get_body_word_counts(submission_id):
    idx = submission_dict['submission_id'].index(submission_id)
    words = word_tokenize(submission_dict['submission_body'][idx])
    return len(words)
```

```python
s_count, t_count = 0, 0
for submission_id in submission_dict['submission_id']:
    s_count += get_title_word_counts(submission_id)
    s_count += get_body_word_counts(submission_id)
print(s_count, "words in the submissions")
for thread_id in delta_thread_dict.keys():
    for comment_dict in delta_thread_dict[thread_id]:
        comment = comment_dict['comment'].replace('\n\n', ' ')
        words = word_tokenize(comment)
        t_count += len(words)
print(t_count, "words in the threads")
print(s_count+t_count, "words in total")
```
